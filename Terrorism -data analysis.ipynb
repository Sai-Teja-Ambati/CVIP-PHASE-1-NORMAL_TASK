# Importing the required libraries

```python
import numpy as np 
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib as mlp
%matplotlib inline
```

# Reading the dataset

```python
dataset = pd.read_csv('D:/coderscave/globalterrorism.csv',encoding='ISO-8859-1')
print("Dataset prepared successfully")
```

```python
#As you can see there are alot of empty values in the datset so we need to select the columns which we need the most
dataset=dataset[["iyear","imonth","iday","country_txt","region_txt","provstate","city",\
       "latitude","longitude","location","summary","attacktype1_txt","targtype1_txt",\
       "gname","motive","weaptype1_txt","nkill","nwound","addnotes"]]
dataset.shape
```

```python
#the first 10 rows of the modified dataset:
dataset.head(10)
```

```python
#Let's rename the columns as they will look pretty
dataset.rename(columns={"iyear":"Year","imonth":"Month","iday":"Day","country_txt":"Country",\
                   "region_txt":"Region","provstate":"Province/State","city":"City",\
                   "latitude":"Latitude","longitude":"Longitude","location":"Location",\
                   "summary":"Summary","attacktype1_txt":"Attack Type","targtype1_txt":"Target Type",\
                   "gname":"Group Name","motive":"Motive","weaptype1_txt":"Weapon Type",\
                   "nkill":"Killed","nwound":"Wounded","addnotes":"Add Notes"},inplace=True)
```

```python
dataset.head(10)
```

```python
dataset.describe(include="all")
```

### As some of the variables are characteristic variables empty is left in some of the columns


```python
dataset.isna().sum()
```

# Still some of the values are left empty 
## As State,City,Summary,Motive,Add notes are characteristic they cannot be assumed by considering mean or anything 
## Latitude,Longitude,Location cannot be assumed 
## but killed wounded can be assumed as 0

## And we are adding a new column damaged to know how many people got damaged in the attack

```python
dataset["Killed"]=dataset["Killed"].fillna(0)
dataset["Wounded"]=dataset["Wounded"].fillna(0)
dataset["Damaged"]=dataset["Killed"]+dataset["Wounded"]
```

# 7.Region Wise Attacks

```python
region=pd.crosstab(dataset.Year,dataset.Region)
region.head()
```

```python
region.plot(kind="area",stacked=False,figsize=(16,12))
plt.title("Region wise attacks")
plt.xlabel("Year")
plt.ylabel("Number of Attacks")
plt.show()
```

```python
regt=region.transpose()
regt["Total"]=regt.sum(axis=1)
ra=regt["Total"].sort_values(ascending=False)
ra
```

```python
ra.plot(kind="bar",figsize=(16,4))
plt.title("Total Number of Attacks in each Region")
plt.xlabel("Region")
plt.ylabel("Number of Attacks")
plt.show()
```
